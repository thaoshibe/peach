```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for token lengths and accuracies
token_lengths = np.array([1, 8, 16, 32, 64, 128, 256])
accuracies = np.array([0.85, 0.92, 1.0, 0.95, 0.96, 0.97, 0.94, 0.93, 0.91])
error = np.array([0.05, 0.08, 0.02, 0.03, 0.03, 0.02, 0.03, 0.04, 0.05])

# Initialize the figure
plt.figure(figsize=(6, 4))

# Plot with error bars
plt.plot(token_lengths, accuracies, 'o-', color='brown', linewidth=1.5, markersize=6)
plt.fill_between(token_lengths, accuracies - error, accuracies + error, color='brown', alpha=0.2)

# Set labels and title
plt.xlabel("Token Length", fontsize=12, weight='bold')
plt.ylabel("Accuracy", fontsize=12, weight='bold')
plt.ylim(0, 1.1)

# Aesthetics for an academic-style plot
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.despine()  # Removes the top and right border for a cleaner look

# Save the figure as a high-quality image for publication
# plt.tight_layout()
# plt.savefig("academic_style_plot.png", dpi=300)
# plt.show()
```
